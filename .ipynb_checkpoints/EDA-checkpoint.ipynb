```python
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine

engine = create_engine("mysql+pymysql://rxuser:RxNorm551!@127.0.0.1:3306/rxnorm")


rel = pd.read_sql("""
SELECT RXCUI1 AS drug_rxcui,
       RXCUI2 AS ingredient_rxcui
FROM rxnrel
WHERE RELA = 'has_ingredient'
LIMIT 500000;
""", engine)

ingredient_count = (
    rel.groupby("drug_rxcui")
       .size()
       .reset_index(name="n_ingredients")
)

plt.figure(figsize=(8,5))
ingredient_count["n_ingredients"].hist(bins=20)
plt.title("Distribution of Ingredient Count per Drug")
plt.xlabel("Number of Ingredients")
plt.ylabel("Number of Drugs")
plt.show()
```

```python
ingredient_freq = (
    rel.groupby("ingredient_rxcui")
       .size()
       .reset_index(name="freq")
       .sort_values("freq", ascending=False)
       .head(15)
)

plt.figure(figsize=(8,5))
plt.bar(range(len(ingredient_freq)), ingredient_freq["freq"])
plt.xticks(range(len(ingredient_freq)), ingredient_freq["ingredient_rxcui"], rotation=90)
plt.title("Top 15 Most Common Ingredients")
plt.ylabel("Frequency")
plt.show()
```

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import mysql.connector
from sklearn.preprocessing import MultiLabelBinarizer
import umap

conn = mysql.connector.connect(
    host="localhost",
    user="rxuser",
    password="RxNorm551!",
    database="rxnorm",
)


QUERY_SMALL = """
SELECT
  r.RXCUI1 AS drug_rxcui,
  d.STR AS drug_name,
  r.RXCUI2 AS ing_rxcui,
  i.STR AS ing_name
FROM rxnrel r
JOIN rxnconso d ON d.RXCUI = r.RXCUI1
JOIN rxnconso i ON i.RXCUI = r.RXCUI2
WHERE r.RELA = 'has_ingredient'
LIMIT 2000
"""

pairs = pd.read_sql(QUERY_SMALL, conn)

print("Rows:", len(pairs))
print(pairs.head())
```

```python
drug_to_ings = (
    pairs.groupby(["drug_rxcui", "drug_name"])["ing_name"]
    .apply(lambda x: list(set(x)))
    .reset_index(name="ingredients")
)

drug_to_ings = drug_to_ings.head(200)

print("Drugs used:", len(drug_to_ings))
```

```python
mlb = MultiLabelBinarizer()
X = mlb.fit_transform(drug_to_ings["ingredients"])

print("Matrix shape:", X.shape)
```

```python
reducer = umap.UMAP(
    n_neighbors=10,
    min_dist=0.3,
    metric="jaccard",
    random_state=42
)

embedding = reducer.fit_transform(X)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(embedding[:,0], embedding[:,1], s=20)
plt.title("Drug Ingredient Similarity (Small Sample)")
plt.show()
```
